```python
import pandas as pd
```

```python
items = pd.read_csv('data/items.csv', sep=';', on_bad_lines='skip')
items.head(5)
```

```python
items.shape
```

```python
order = pd.read_csv('data/orders.csv', sep=';', on_bad_lines='skip')
order.head()
```

```python
# Removing the items where are not present in the order database

items.drop(items.loc[items['item_code'].isin(order.item_code)==False].index, inplace=True)
```

```python
# Replacing item_name with the most common value

items['item_name'] = items['item_name'].fillna(items['item_name'].mode().iloc[0])
```

```python
# deleting column['group2_id'] and ['group2'] as they have many NA values

items = items.drop(columns=['group2_id','group2'])
```

```python
items = items.drop(columns=['group0_id','group1_id'])
```

```python
# Filling NA values of group0 & group1 with most common values

items['group0'] = items['group0'].fillna(items['group0'].mode().iloc[0])
items['group1'] = items['group1'].fillna(items['group1'].mode().iloc[0])
```

```python
# Filling NA values of gender, age, style, size, color and category with most common values

items['gender'] = items['gender'].fillna(items['gender'].mode().iloc[0])
items['age'] = items['age'].fillna(items['age'].mode().iloc[0])
items['color'] = items['color'].fillna(items['color'].mode().iloc[0])
items['size'] = items['size'].fillna(items['size'].mode().iloc[0])
items['style'] = items['style'].fillna(items['style'].mode().iloc[0])
items['category'] = items['category'].fillna(items['category'].mode().iloc[0])
```

```python
items['size'] = items['size'].astype(str)
```

```python
items.isna().sum()
```

```python
items.to_csv('items_cleaned.csv', index=False)
```
