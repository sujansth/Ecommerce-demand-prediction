```python
import pandas as pd
```

```python
order = pd.read_csv('data/order_cleaned.csv')
order.head()
```

```python
items = pd.read_csv('data/items_cleaned.csv')
items.head()
```

```python
items.shape
```

```python
df = order.merge(items, on='item_code')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
df = df.drop(columns=['order_id','item_code','owner','channel','department','brand_id','CreatedAt','style','unit_cogs','unit_rrp_vat_excl'])
```

```python
df['total_revenue'] = df['quantity'] * df['unit_price_vat_excl']
```

```python
df = df.drop(columns=['quantity','unit_price_vat_excl'])
```

```python
df = df.rename(columns={'name':'brand_name'})
```

```python
df = df.sort_values('date')
```

```python
df.to_csv('final_data.csv', index=False)
```
