```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/final_data.csv')
```

```python
df
```

```python
top_group0_rev = pd.DataFrame(df.groupby('group0')['total_revenue'].sum().sort_values(ascending=False))
top_group0_rev['total_revenue'] = round(top_group0_rev['total_revenue'] / 1000, 2)
top_group0_rev
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(top_group0_rev.index, top_group0_rev.total_revenue)
plt.xticks(rotation=60, size=15)
plt.xlabel("Top 10 groups", fontsize=15)
plt.ylabel("Sales Revenue (in thousands)", fontsize=20)
plt.title("Top 10 groups", fontsize=30)
plt.show()
```

```python
top_group1_rev = pd.DataFrame(df.groupby('group1')['total_revenue'].sum().sort_values(ascending=False)).head(10)
top_group1_rev['total_revenue'] = round(top_group1_rev['total_revenue'] / 1000, 2)
top_group1_rev
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(top_group1_rev.index, top_group1_rev.total_revenue)
plt.xticks(rotation=60, size=15)
plt.xlabel("Top 10 sub-groups", fontsize=15)
plt.ylabel("Sales Revenue (in thousands)", fontsize=20)
plt.title("Top 10 sub-groups", fontsize=30)
plt.show()
```

```python
top_site = pd.DataFrame(df.groupby('site')['total_revenue'].sum().sort_values(ascending=False).head(10))
top_site['total_revenue'] = round(top_site['total_revenue'] / 1000, 2)
top_site
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(top_site.index, top_site.total_revenue)
plt.xticks(rotation=60, size=15)
plt.xlabel("Top 10 websites", fontsize=15)
plt.ylabel("Sales Revenue (in thousands)", fontsize=20)
plt.title("Top 10 selling sites for Top4Sport", fontsize=30)
plt.show()
```

```python
top_brand = pd.DataFrame(df.groupby('brand_name')['total_revenue'].sum().sort_values(ascending=False)).head(10)
top_brand['total_revenue'] = round(top_brand['total_revenue'] / 1000, 2)
top_brand
```
